```python
import matplotlib.pyplot as plt
import numpy as np


fig = plt.figure()
ax = fig.add_subplot(projection='3d')
ax.set_axis_off()

# Make data
u = np.linspace(0, 2 * np.pi, 100)
v = np.linspace(0, np.pi /4, 100)
x = np.outer(np.cos(u), np.sin(v))
y = np.outer(np.sin(u), np.sin(v))
z = np.outer(np.ones(np.size(u)), np.cos(v))

# Plot the surface
ax.plot_surface(x, y, z, alpha=0.2)

# Set an equal aspect ratio
ax.set_aspect('equal')

import numpy as np
phi = [np.radians(12)]
N_examples = 5
theta = np.radians(np.arange(0, 360, 30))#[i * 2 * np.pi/(N_examples - 1) for i in range(1, N_examples)]

phi, theta = np.meshgrid(phi, theta)

ax.scatter3D(
    xs=np.cos(theta)*np.sin(phi), 
    ys=np.sin(theta)*np.sin(phi),
    zs=np.cos(phi),
    c='r',
    label='Vision sensor'
)

ax.scatter3D(xs=[0], ys=[0], zs=[0], c='black')
ax.text(0, 0.15, 0, 'contact point')

plt.legend(bbox_to_anchor=(4, 1), loc="upper left")


# Put a legend to the right of the current axis
ax.legend(loc='center left', bbox_to_anchor=(1, 0.5))
```
